# Random Forest

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier
```

```python
db=pd.read_csv(r"C:\Users\H S COMPUTER S HYD\Desktop\\diabetes_prediction_dataset.csv")
```

```python
db.head()
```

```python
db.drop(["smoking_history","gender"],axis=1,inplace=True)
```

```python
db.head()
```

```python
x=db.iloc[:,:-1]
y=db.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
df=RandomForestClassifier()
```

```python
df.fit(x_train,y_train)
```

```python
test_pred=df.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,precision_score,f1_score,recall_score,confusion_matrix,classification_report
```

```python
print(f"test accuracy: ",{accuracy_score(y_test,test_pred)})
print("Classification Report")
print(f"{classification_report(y_test,test_pred)}")
```

```python
from sklearn.model_selection import train_test_split
```

```python
dt=DecisionTreeClassifier()
```

```python
dt.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score,precision_score,f1_score,recall_score,confusion_matrix,classification_report
```

```python
test_pred=dt.predict(x_test)
```

```python
print(f"test accuracy: ",{accuracy_score(y_test,test_pred)})
print(f"{classification_report(y_test,test_pred)}")
```
