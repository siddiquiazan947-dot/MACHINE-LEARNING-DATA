# Classification

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
```

```python
iris=sns.load_dataset("iris")
```

```python
iris=iris[iris.species != "virginica"]
```

```python
iris.species.unique()
```

```python
iris["species"]=iris.species.map({'setosa':0, 'versicolor':1})
```

```python
iris
```

```python
iris.drop("sepal_width",axis=1,inplace=True)
```

```python
x=iris.iloc[:,:-1]
```

```python
y=iris.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
lr_classifier=LogisticRegression()
```

```python
lr_classifier.fit(x_train,y_train)
```

```python
y_test_pred=lr_classifier.predict(x_test)
```

```python
y_test_pred
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix
```

```python
sns.heatmap(confusion_matrix(y_test,y_test_pred),annot=True)
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
```

```python
print("test accuracy:",accuracy_score(y_test,y_test_pred))
```

```python
print("recall:",recall_score(y_test,y_test_pred))
```

```python
print("precision:",precision_score(y_test,y_test_pred))
```

```python
print("f1:",f1_score(y_test,y_test_pred))
```

```python
print(confusion_matrix(y_test,y_test_pred))
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_test_pred))
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
print(mean_absolute_error(y_test,y_test_pred))
```

```python
print(r2_score(y_test,y_test_pred))
```

* HYPER PARAMETER TUNNING

```python
from sklearn.model_selection import GridSearchCV
```

```python
data_Classifier=LogisticRegression()
```

```python
parm_grid={"C":[1,2,4,5,6,9,10],"tol":[0.001,0.0001,0.01]}
```

```python
cancer_grid=GridSearchCV(estimator=data_Classifier,param_grid=parm_grid,cv=5,scoring="accuracy")
```

```python
cancer_grid.fit(x_train,y_train)
```

```python
cancer_grid.best_params_
```

```python
cancer_grid.best_score_
```

```python
c_predict=cancer_grid.predict(x_test)
```

```python
c_predict
```

```python
print("test accuracy:",accuracy_score(y_test,c_predict))
```

```python
print(classification_report(y_test,c_predict))
```
