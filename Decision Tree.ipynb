```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

# Decision Tree 

* 01.Entropy=H(s)=-(p+)log2(p+)-(p-)log2(p-)
* 02.Ginni Impurity=G.I=1-(p)^2

```python
iris=sns.load_dataset("iris")
```

```python
iris.species.unique()
```

```python
iris["species"]=iris.species.map({'setosa':0, 'versicolor':1, 'virginica':2})
```

```python
iris.head(5)
```

```python
x=iris.iloc[:,:-1]
y=iris.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
dt=DecisionTreeClassifier()
```

```python
dt.fit(x_train,y_train)
```

```python
test_pred=dt.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,precision_score,f1_score,recall_score,confusion_matrix,classification_report
```

```python
print(f"test accuracy: ",{accuracy_score(y_test,test_pred)})
print(f"{classification_report(y_test,test_pred)}")
```

```python
peg=sns.load_dataset("penguins")
```

```python
peg=peg.dropna()
```

```python
peg.sex.unique()
peg.island.unique()
peg.species.unique()
```

```python
peg["sex"]=peg.sex.map({'Male':0,'Female':1})
```

```python
peg["island"]=peg.island.map({'Torgersen':0, 'Biscoe':1, 'Dream':2})
```

```python
peg["species"]=peg.species.map({'Adelie':0, 'Chinstrap':1, 'Gentoo':2})
```

```python
x=peg.iloc[:,1:]
y=peg.iloc[:,1]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
pg=DecisionTreeClassifier()
```

```python
pg.fit(x_train,y_train)
```

```python
png_pred=pg.predict(x_test)
```

```python
print(f"test accuracy: ",{accuracy_score(y_test,png_pred)})
print(f"{classification_report(y_test,png_pred)}")
```

```python
plt.figure(figsize=(10,7))
tree.plot_tree(pg,filled=True)
```
