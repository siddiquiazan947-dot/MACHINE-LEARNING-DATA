```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_blobs
```

```python
x,y = make_blobs(n_samples=300, centers=3, random_state=42)
```

```python
plt.scatter(x[:,0],x[:, 1] ,c =y)
```

```python
import scipy.cluster.hierarchy as sc
```

```python
sc.dendrogram(sc.linkage(x,method='ward'))
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
cluster = AgglomerativeClustering(n_clusters=3,linkage='ward')
```

```python
labels = cluster.fit_predict(x)
```

```python
plt.scatter(x[:,0],x[:,1], c = labels)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris().data
```

```python
plt.scatter(iris[:,0],iris[:,1])
```

```python
scaler  = StandardScaler()
iris_scaled = scaler.fit_transform(iris)
```

```python
iris_scaled
```

```python
iris_cluster_agg = AgglomerativeClustering(n_clusters=3,linkage='ward')
```

```python
iris_labels = iris_cluster_agg.fit_predict(iris_scaled)
```

```python
plt.scatter(iris[:,0],iris[:,1], c=iris_labels)
```
