# DB SCAN

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN,KMeans,AgglomerativeClustering
```

```python
scaler=StandardScaler()
x,y=make_moons(n_samples=500,noise=0.05)
```

```python
x_scaled=scaler.fit_transform(x)
x_scaled.shape
```

```python
plt.scatter(x[:,0],x[:,1])
```

```python
db_clustering=DBSCAN(min_samples=4,eps=0.5)
```

```python
db_clustering.fit(x_scaled)
```

```python
plt.scatter(x_scaled[:,0],x_scaled[:,1],c=db_clustering.labels_)
```

## K-Mean

```python
kmeans=KMeans(n_clusters=2)
```

```python
kmeans.fit(x_scaled)
```

```python
plt.scatter(x_scaled[:,0],x_scaled[:,1],c=kmeans.labels_)
```

## AgglomerativeClustering

```python
Agg_cluster=AgglomerativeClustering(n_clusters=2,linkage="ward")
```

```python
Agg_cluster.fit(x_scaled)
```

```python
plt.scatter(x_scaled[:,0],x_scaled[:,1],c=Agg_cluster.labels_)
```

```python
from sklearn.metrics import silhouette_score
```

```python
print("silhouette_score K-Means:",silhouette_score(x_scaled,kmeans.labels_))
```

```python
print("silhouette_score AgglomerativeClustering Score:",silhouette_score(x_scaled,Agg_cluster.labels_))
```

```python
print("silhouette_score DB Score:",silhouette_score(x_scaled,db_clustering.labels_))
```

# TASK

```python
from sklearn.datasets import load_breast_cancer
```

```python
data=load_breast_cancer()
```

```python
x=data.data
```

```python
scaler=StandardScaler()
```

```python
x_scale=scaler.fit_transform(x)
x_scale.shape
```

```python
plt.scatter(x_scale[:,0],x_scale[:,1],c=data.target)
```

# DB SCAN

```python
db_clustering=DBSCAN(min_samples=4,eps=0.5)
```

```python
db_clustering.fit(x_scale)
```

```python
plt.scatter(x_scale[:,0],x_scale[:,1],c=db_clustering.labels_)
```

# K-Mean

```python
kmean=KMeans(n_clusters=2)
```

```python
kmean.fit(x_scale)
```

```python
plt.scatter(x_scale[:,0],x_scale[:,1],c=kmean.labels_)
```

# AgglomerativeClustering

```python
Agg_cluster=AgglomerativeClustering(n_clusters=2,linkage="ward")
```

```python
Agg_cluster.fit(x_scale)
```

```python
plt.scatter(x_scale[:,0],x_scale[:,1],c=Agg_cluster.labels_)
```

```python
print("silhouette_score K-Means:",silhouette_score(x_scale,kmean.labels_))
```

```python
print("silhouette_score AgglomerativeClustering Score:",silhouette_score(x_scale,Agg_cluster.labels_))
```
