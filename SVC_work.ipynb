# Support Vector Machine.

```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
from sklearn.svm import SVC
```

```python
iris=load_iris()
Data=pd.DataFrame(iris.data,columns=iris.feature_names)
```

```python
Data
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(Data,iris.target,test_size=0.2,random_state=42)
```

```python
svc=SVC(C=5)
```

```python
svc.fit(x_train,y_train)
```

```python
x_pred=svc.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(y_test,x_pred))
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
br_cancer=load_breast_cancer()
Data=pd.DataFrame(br_cancer.data,columns=br_cancer.feature_names)
```

```python
Data
```

```python
x_train,x_test,y_train,y_test=train_test_split(Data,br_cancer.target,test_size=0.2,random_state=42)
```

```python
svc=SVC(C=5)
```

```python
svc.fit(x_train,y_train)
```

```python
x_pred=svc.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
```

```python
print(f"accuracy_score:",accuracy_score(y_test,x_pred))
```

```python
print(f"precision_score:",precision_score(y_test,x_pred))
```

```python
print(f"recall_score:",recall_score(y_test,x_pred))
```

```python
print(f"f1_score:",f1_score(y_test,x_pred))
```
